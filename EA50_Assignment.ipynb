## Final Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
plt.title('Comparison of Summer Temperatures\n and Specimen Collection Dates over time')

color = 'tab:red'
sns.lineplot(x=merged_p_zeliacon_df['Year'], y=merged_p_zeliacon_df['BC Summer Temp (˚C)'], color=color)
ax1.set_xlabel('Year')
ax1.set_ylabel('BC Summer Temp (˚C)', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Days of Specimen Collections', color=color)
sns.regplot(x=merged_p_zeliacon_df['Year'], y=merged_p_zeliacon_df['Day of Year'], color=color, fit_reg = False, x_jitter = 0.8, y_jitter = 0.8, scatter_kws = {'alpha' : 1/2})
ax2.set_ylabel('Day of Specimen Collection', color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.tight_layout()
plt.show()
```
